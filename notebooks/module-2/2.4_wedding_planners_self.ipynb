```python
from dotenv import load_dotenv

load_dotenv()
```

```python
# This notebook demonstrates how to create a multi-agent system where a main agent can call multiple subagents, each with its own model and tools. The main agent will decide which subagent to call based on the user's query.

# Step 1: Initialize the models for the main agent and subagents
from langchain.chat_models import init_chat_model


model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")
model2 = init_chat_model("llama-3.1-8b-instant", model_provider="groq")
```

```python
from langchain_mcp_adapters.client import MultiServerMCPClient

client = MultiServerMCPClient(
    {
        "travel_server": {
                "transport": "streamable_http",
                "url": "https://mcp.kiwi.com"
            }
    }
)

tools = await client.get_tools()
```

```python
# create subagent to look for flights for wedding
from langchain.agents import create_agent

subagent_1 = create_agent(
    model=model,
    tools=[tools]
)
```
