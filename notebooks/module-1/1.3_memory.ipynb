```python
from dotenv import load_dotenv

load_dotenv()
```

## No memory

```python
from langchain.agents import create_agent

from langchain.chat_models import init_chat_model

model = init_chat_model(model="llama-3.1-8b-instant", model_provider="groq" , temperature=0)

agent = create_agent(
    model=model
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Chiranjeev and my favourite colour is violet. What is my name and favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="What is my Name, and What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    model=model,
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What is my Name, and What's my favourite colour?")

config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
 
pprint(response)
```
